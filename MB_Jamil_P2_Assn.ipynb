### Part II: Scraping the Katz School’s AI Faculty Web Page

```python
# load the necessary libraries for webscrapping
from bs4 import BeautifulSoup
import requests
import csv


# defining the Katz School's website:
url = 'https://www.yu.edu/katz/programs/graduate/artificial-intelligence/faculty'
source = requests.get(url).text

# creating a variable soup to parse our source(the webpage we downloaded):
soup = BeautifulSoup(source, 'lxml')
```

```python
# Now locating the body, and saving it in a separate variable:
faculty = soup.find('div', class_='body')

#printing the results:
print(faculty)
```

We see from our body that most of the elements are of either 'p' tag or 'h5' tag. I would like to explain here how these two are different:
1. 'p' tag elements defines a paragraph and they are usually represented by lines of texts in a html webpage.
2. 'h5' tag elements defines a 5th level heading in a html webpage.

```python
faculty_list = []
```

```python
'''Now, since we know which elements we will search and put in our dataframe. We will create a new variable where we will find
the data represented by the tags "p" and "h5"'''

search = faculty.find_all(['h5','p'])
```

```python
# Now from our body, we also see that every faculty member from the Katz School have 2 'p' tags and 1 'h5' tag. So it means
# that every faculty member has a total of '3' tags. Now, we will define a function:


''' We will now use a for loop, which will loop over our search variable and will search for text, we will also create an
empty dictionary which will save the results of name, title by using a (,) comma since we know in the body name and title 
are separated by a comma and is a h5 tag. 

Then, the bio and teaching background is a paragraph so we will pick that up with increasing our index and then append 
them into our dictionary as well. At the end, we wil append our dictionary results to the faculty_list which we earlier.'''

for i in range(0, len(search), 3):
    faculty_dict = {}
    header = search[i].text  
    faculty_dict['Name'], faculty_dict['Title'] = header.split(',', 1)
    faculty_dict['Bio'] = search[i+1].text
    faculty_dict['Teaching_Background'] = search[i+2].text
  # To store our faculty_dictionary to our faculty_list:
    faculty_list.append(faculty_dict)
```

```python
# let's view our faculty_list:
faculty_list
```

* It seems everything looks great, and is as per the website as well. So will now create the dataframe with this list:

```python
faculty_info = pd.DataFrame(faculty_list)
```

```python
#Let's now view our dataframe:
faculty_info
```

* Hence, we see in our result, that the webpage is now successfully saved into a readable dataframe and can be accessed easily with respect to index for more detail reading.
